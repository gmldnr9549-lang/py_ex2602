```python
print("예제1")
s = "안녕하세요"

print("==s길이 출력 ==")
print(len(s))

#s[start:end(-1):step]
print("s를 이루는 문자 중에서 앞에 거 2개 출력")
print(s[0:2])

print("s를 이루는 문장 중에서 문자 4개출력, 첫 인덱스 생략")
print(s[:4])

print("s를 이루는 문장 중에서 압엣 문자 3개 건너 뛰고, 나머지 전부 출력, len")
print(s[3:])

print("==예제3==")
s = "a안b녕c하d세e요f"

print("s를 이루는 문장 중에서 한글만 출력")
print(s[1] +s[3] +s[5] +s[7] +s[9])

print("s를 이루는 문장 중에서 한글만 출력, for, if, or, continue, 방법 1")
for c in s:
    if c == 'a' or c == 'b' or c == 'c' or c == 'd' or c == 'e' or c == 'f':
        continue
    print(c)
```
