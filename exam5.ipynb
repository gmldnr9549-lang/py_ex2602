```python
import re

text = "안녕하세요 저는 15살 입니다, 저는 안경이 15개 있습니다"

#\d는 숫자
# +는 1개 이상

pattern = r"\d+"
result = re.findall(pattern, text)
print(result)
```

```python
sample = "abc123def456"

re.findall(r"\d+", sample)

re.findall(r"[a-z]+", sample)
```
